```python
# read raw file
import numpy as np
with open('/root/project/scripts/eval_scripts/reconstructed_volume.raw', 'rb') as f:
    raw_data = np.frombuffer(f.read(), dtype=np.float32)
    raw_data = raw_data.reshape((128, 128, 128))  # reshape to (Z, Y, X)
```

```python
import matplotlib.pyplot as plt
# visualize a slice of the volume
slice_index = 25  # choose a slice index along the Z-axis
plt.imshow(raw_data[slice_index, :, :], cmap='gray')
plt.title(f'Slice {slice_index} of Reconstructed Volume')
plt.colorbar()
plt.show()
```
